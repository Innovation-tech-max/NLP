##  **Lexical richness: Number of unique words used in a text.** 

```python
from __future__ import division
```

```python
a=len(text1)/len(set(text1))
print("Average usage time of each word is",a)
```

```python
text3.count("smote")
b=100*text3.count("smote")/len(text3)
print("Percentage of occurence of the given word is",b,"%")
```

```python
text5.count("lol")
c=100*text5.count("lol")/len(text5)
print("Percentage of occurence of the given word is",c,"%")
```

```python
def lexical_diversity(text):
  return len(text) / len(set(text))


def percentage(count, total):
  return 100 * count / total
```

```python
percentage(text4.count('a'), len(text4))
```
